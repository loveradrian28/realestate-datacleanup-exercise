#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
#Importando la biblioteca adicional requerida
from scipy.stats import zscore

#Normalizacion de precios usando zscore
CS = CS.copy()
CS.loc[:, 'precio_normalizado'] = CS.groupby('level5')['price'].transform(zscore)

#Codigo para generar histograma
plt.figure(figsize=(10, 6))

poblaciones = CS['level5'].unique()

for x in poblaciones:
    precios_norm = CS[CS['level5'] == x]['precio_normalizado']
    plt.hist(precios_norm, bins=10, alpha=0.5, label=f'Población: {x}', density=True)

plt.title('Distribución de Precios Normalizados por Población')
plt.xlabel('Precio Normalizado (Z-score)')
plt.ylabel('Densidad')
plt.legend()

plt.tight_layout()
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)
 
Hint: Make a scatter plot of each population using subplots.

```python
#Codigo para crear la figura y los subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle('Gráficos de Dispersión por Población', fontsize=16)

poblaciones = CS['level5'].unique()

for i, poblacion in enumerate(poblaciones):
    row = i // 2
    col = i % 2
    
    data = CS[CS['level5'] == poblacion]
    
    #Creación del diagrama de dispersión
    axs[row, col].scatter(data['surface'], data['price'], alpha=0.6)
    
    #Personalización de la figura
    axs[row, col].set_title(f'Población: {poblacion}')
    axs[row, col].set_xlabel('Superficie (m²)')
    axs[row, col].set_ylabel('Precio (€)')
    axs[row, col].grid(True)

plt.tight_layout()
plt.show()
```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
from ipyleaflet import Map, basemaps

# Mapa centrado en (60 grados latitud y -2.2 grados longitud)
# Latitud, longitud
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:
#Importando bibliotecas adicionales
from ipyleaflet import MarkerCluster, CircleMarker, AwesomeIcon, Marker

#Diccionario de las columnas 'latitude' y 'longitude' del dataframe CS
coordenadas = CS.groupby('level5')[['latitude', 'longitude']].apply(lambda x: x.to_dict(orient='records')).to_dict()

#Asignación de colores para cada población
colores = {
    'Getafe': 'red',
    'Alcorcón': 'blue',
    'Leganés': 'green',
    'Fuenlabrada': 'purple'
}

#Centrando mapa a la localidad de Madrid, España
mapa = Map(center=(40.4168, -3.7038), zoom=10)

#Creacion de marcadores para las coordenadas
marcadores = []

for level5, coordenadas in coordenadas.items():
    color = colores.get(level5,'gray')  # Default color si no se encuentra en el diccionario
    for coord in coordenadas:
        marcador = CircleMarker(
            location=(coord['latitude'], coord['longitude']),
            radius=5,
            color=color,
            fill_color=color,
            fill_opacity=0.7
        )
        marcadores.append(marcador)

for marcador in marcadores:
    mapa.add_layer(marcador)

#Ubicacion de poblaciones de interés
legend_locations = {
    'Getafe': (40.3, -3.7),
    'Alcorcón': (40.35, -3.8),
    'Leganés': (40.32, -3.75),
    'Fuenlabrada': (40.28, -3.8)
}

for poblacion, color in colores.items():
    icono = AwesomeIcon(name='circle', marker_color=color, icon_color='white')
    marcador_leyenda = Marker(icon=icono, location=legend_locations[poblacion])
    mapa.add_layer(marcador_leyenda)

mapa
```
